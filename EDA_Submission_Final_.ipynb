## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
def home_type(reviews):
  if reviews>100:
    return "Excellent"
  elif reviews<100 and reviews>50:   #define the function for giving the review and you can easily analyse by checking review
    return "Good"
  else:
    return "Avg"
df_air["remark"]=df_air.apply(lambda x:home_type(x['number_of_reviews']),axis=1)  # add review column for easy anlysis
df_air.head()
````

**By this you can analyse by checking the remark because on the basis of number of review of any home  **

#### Chart - 5

**Price at different location in New York**

````python
# Chart - 5 visualization code
area_price=df_air.groupby(["neighbourhood_group","remark"])["price"].mean().reset_index(name="avg_price")
area_price.head()
````

````python
sns.barplot(data=area_price,x='neighbourhood_group',y='avg_price',hue="remark")
plt.xlabel("Different area",weight="bold",color="r")
plt.ylabel("Avg price",weight="bold",color="r")
plt.title("Prices of different area according to remark",weight="bold",color="g")
plt.show()
````

````python
def demand(availbility):        # define the function for adding one column which name is demand
  if availbility>200:           # this function work according to availability of homes
    return "Low"
  elif availbility<200 and availbility>100:
    return "Medium"
  elif availbility<100 and availbility>50:
    return "High"
  else:
    return "Very High"
df_air["demand"]=df_air['availability_365'].apply(demand)
df_air.head()
````

````python
df_demand=df_air.groupby(['neighbourhood_group',"demand"],as_index=True).agg({"price":"mean",'minimum_nights':"mean",'number_of_reviews':"mean",'availability_365':"mean"}).sort_values(by="price",ascending=False).rename(columns={"price":"avg_price",'number_of_reviews':"avg_number_of_reviews",'minimum_nights':'avg_minimum_nights'}).reset_index()
sns.barplot(data=df_demand,x='neighbourhood_group',y='avg_price',hue="demand")
plt.xlabel("Different area",weight="bold",color="r")
plt.ylabel("Avg price",weight="bold",color="r")
plt.title("Prices of different area according to demand",weight="bold",color="g")
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose this graph because it is catagrical data.It is the best way to represent avg price  with respect to remark and demand.

I found that avg price of homes the lowest which remark is excellent and which homes demand is very high it means that avg price is inversely proptional to remark and demand because homes  which provide good services and have good reviews are also pocket friendly that's people likes that homes.
